```
import pandas as pd
import os

def append_to_metadata(metadata, metadata_csv_path):
    """
    Appends metadata to an existing metadata.csv file, or creates a new one if it doesn't exist.

    Parameters:
    metadata (list): List of dictionaries containing metadata.
    metadata_csv_path (str): Path to the metadata CSV file.

    Returns:
    None
    """
    # Check if metadata.csv already exists
    if os.path.exists(metadata_csv_path):
        # Load existing metadata
        existing_metadata = pd.read_csv(metadata_csv_path)
        # Concatenate new metadata with existing
        combined_metadata = pd.concat([existing_metadata, pd.DataFrame(metadata)], ignore_index=True)
    else:
        # No existing metadata, use the new metadata
        combined_metadata = pd.DataFrame(metadata)

    # Save back to metadata.csv
    combined_metadata.to_csv(metadata_csv_path, index=False)
```

```
from pydub import AudioSegment
import os
import pandas as pd

# Paths for input and output
input_folder = "/content/extracted_data/recordings_b/1"  # Update this to your input folder
output_folder = "/content/wavdata/WAV_Files"  # Update this to your output folder
metadata_csv_path = "/content/metadata.csv"  # Metadata file path

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Convert MP3 to WAV and collect metadata
metadata = []
for file_name in os.listdir(input_folder):
    if file_name.endswith(".mp3"):
        mp3_path = os.path.join(input_folder, file_name)
        wav_file_name = file_name.replace(".mp3", ".wav")
        wav_path = os.path.join(output_folder, wav_file_name)

        try:
            # Convert and save
            audio = AudioSegment.from_mp3(mp3_path)
            audio.export(wav_path, format="wav")

            # Append metadata
            metadata.append({"file_name": wav_file_name, "label": ""})  # Add label if needed
        except Exception as e:
            print(f"Error processing {file_name}: {e}")

# Append metadata to the main CSV file
append_to_metadata(metadata, metadata_csv_path)
print(f"Processed {len(metadata)} files. Metadata updated at {metadata_csv_path}")
```

**Load Audio Dataset**

```
import os
from datasets import Dataset, Audio

# Paths for dataset and metadata
DATASET_PATH = "/content/wavdata/WAV_Files"
METADATA_FILE = "/content/wavdata/WAV_Files/metadata.csv"

def load_audio_dataset(dataset_path, metadata_file=None):
    if metadata_file:
        dataset = Dataset.from_csv(metadata_file)
        dataset = dataset.filter(lambda x: os.path.exists(os.path.join(dataset_path, x["file_name"])))
        dataset = dataset.map(lambda x: {"file_name": os.path.join(dataset_path, x["file_name"])})
    else:
        files = [{"file_name": os.path.join(dataset_path, f)} for f in os.listdir(dataset_path) if f.endswith(".wav")]
        if not files:
            raise FileNotFoundError("No .wav files found!")
        dataset = Dataset.from_list(files)

    dataset = dataset.cast_column("file_name", Audio(sampling_rate=16000))
    return dataset

audio_dataset = load_audio_dataset(DATASET_PATH, METADATA_FILE)
print(f"Dataset loaded with {len(audio_dataset)} samples.")
```

****Feature Extraction with Wav2Vec2:****

```
from transformers import Wav2Vec2Processor, Wav2Vec2Model
import torch

processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-large-960h")
teacher_model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-large-960h")
teacher_model.eval()

def extract_teacher_features(batch):
    audio_arrays = [audio["array"] for audio in batch["file_name"]]
    inputs = processor(audio_arrays, sampling_rate=16000, return_tensors="pt", padding=True)
    with torch.no_grad():
        features = teacher_model(inputs.input_values).last_hidden_state.mean(dim=1).numpy()
    batch["teacher_features"] = features
    return batch
print("Feature extraction completed.")

audio_dataset = audio_dataset.map(extract_teacher_features, batched=True)
print("Feature extraction completed.")

```

**Train the student model wav2vec2-base:**

```
from transformers import Wav2Vec2ForSequenceClassification, TrainingArguments, Trainer

student_model = Wav2Vec2ForSequenceClassification.from_pretrained(
    "facebook/wav2vec2-base", num_labels=2  # Adjust labels as per your dataset
)

def preprocess_for_student(batch):
    return {
        "input_values": batch["teacher_features"],
        "labels": batch["labels"]  # Ensure labels are provided
    }

processed_dataset = audio_dataset.map(preprocess_for_student, remove_columns=audio_dataset.column_names)

training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    num_train_epochs=10,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=10,
)

trainer = Trainer(
    model=student_model,
    args=training_args,
    train_dataset=processed_dataset,
    eval_dataset=processed_dataset,
)

trainer.train()
```
